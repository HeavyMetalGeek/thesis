### Process probe data ###

```python
def process_probes(probe_path):
    os.system('sed -i_orig \'s/[(,)]//g\' {}'.format(probe_path))
    u = pd.read_csv(probe_path, 
                    delim_whitespace=True, 
                    comment='#', 
                    header=None, 
                    index_col=0)
    num_pts = int(u.shape[1]/3)
    cols = pd.MultiIndex.from_tuples([(x,y) for x in range(num_pts) 
                                      for y in ['ux', 'uy', 'uz']])

    u.columns = cols
    u.index.name = 'Time'
    u.columns.names = ['Probes', 'Component']
    return u
```

```python
### Plot Configuration ###
plot_config()

### Processing Parameters ##
# Number of points to process at tail of data #
N = 8000
# Time step
dt = 0.00025
# t0 index
#Nt0 = -N*2-1

### Set which simulation to process ###
# 100 for fine, 60 for coarse
u = None
sim = 100
if (sim == 100):
    u = process_probes('./postProcessing/center_probe/290.5/U')
else:
    u = process_probes('./postProcessing_60/center_probe/50/U')

#u = u.iloc[Nt0:]
ueff, up = u, u-u.mean()
u_orig = u

u = np.array(ueff[0]['ux'])
v = np.array(ueff[0]['uy'])
w = np.array(ueff[0]['uz'])
fx = np.array(up[0]['ux'])
fy = np.array(up[0]['uy'])
fz = np.array(up[0]['uz'])
```

### Calculating velocity autocorrelation using Fourier transform ###
$\displaystyle u' = \tilde{u} - U$

$\displaystyle R_{xx}(m) = \lim_{T\to\infty} \left( \frac{1}{T} \right) \int_0^T f(t) \, f^*(t) \, dt$

$\displaystyle R_{xx}(m) = \mathcal{F}_x \left[ f_{\nu,i} \; f^*_{\nu,i} \right]$

```python
#Nx = len(fx)
Nx = N
fvix = np.fft.fft(fx, n=2*Nx)
#fvix = np.fft.fft(fx[-N:])
acfx = fvix * np.conjugate(fvix)
acfx = np.fft.ifft(acfx)
#filter out positive lags
acfx = acfx[:Nx]
#filter out real part
acfx = np.real(acfx)
acfx = acfx/acfx[0]

#Ny = len(fy)
Ny = N
fviy = np.fft.fft(fy, n=2*Ny)
#fviy = np.fft.fft(fy[-N:])
acfy = fviy * np.conjugate(fviy)
acfy = np.fft.ifft(acfy)
#filter out positive lags
acfy = acfy[:Ny]
#filter out real part
acfy = np.real(acfy)
acfy = acfy/acfy[0]

#Nz = len(fz)
Nz = N
fviz = np.fft.fft(fz, n=2*Nz)
#fviz = np.fft.fft(fz[-N:])
acfz = fviz * np.conjugate(fviz)
acfz = np.fft.ifft(acfz)
#filter out positive lags
acfz = acfz[:Nz]
#filter out real part
acfz = np.real(acfz)
acfz = acfz/acfz[0]
```

### Calculating velocity autocorrelation using discrete values ###

$\displaystyle \hat{R}_{xx}(m) = \left( \frac{1}{N} \right) \sum_{i=0}^{N-m-1} f_i \, f^*_{i+m}$

```python
Rxx = []
Ryy = []
Rzz = []

#lag = list(range(0, N))
#fxi = np.array(fx.iloc[:N])
#fyi = np.array(fy.iloc[:N])
#fzi = np.array(fz.iloc[:N])
fxi = fx[:N]
fyi = fy[:N]
fzi = fz[:N]

def autocorr(obj, N, lag):
    obji = obj[-(N+lag):][:N]
    objim = obj[-N:]
    acov = obji * objim
    var = obji**2
    return acov.sum() / var.sum()

m = N
for i in range(m):
    Rxx.append(autocorr(fx, N, i))
    Ryy.append(autocorr(fy, N, i))
    Rzz.append(autocorr(fz, N, i))

# for m in lag:
#     fxim = np.array(fx.iloc[Nt0+m:].iloc[:N])
#     Rx = fxi * fxim
#     fxvar = (fxi**2).sum()/N
#     Rxx.append(Rx.sum()/N/fxvar)
    
#     fyim = np.array(fy.iloc[Nt0+m:].iloc[:N])
#     #Ry = fyi * fyim.conj()
#     #Ryy.append(Ry.sum()/N)
#     Ry = fyi * fyim
#     fyvar = (fyi**2).sum()/N
#     Ryy.append(Ry.sum()/N/fyvar)
    
#     fzim = np.array(fz.iloc[Nt0+m:].iloc[:N])
#     #Rz = fzi * fzim.conj()
#     #Rzz.append(Rz.sum()/N)
#     Rz = fzi * fzim
#     fzvar = (fzi**2).sum()/N
#     Rzz.append(Rz.sum()/N/fzvar)
```

```python
fname = './postProcessing/center_probe/290.5/U'
u = pd.read_csv(fname,delim_whitespace=True,comment='#',header=None,index_col=0)
cols = pd.MultiIndex.from_tuples([(x,y) for x in range(int(u.shape[1]/3)) for y in ['ux','uy','uz']])
u.columns = cols
u.index.name = 'Time'
u.columns.names = ['Probes','Component']
ueff = u
up = ueff-ueff.mean()

uheader = pd.read_csv(fname,
                      delim_whitespace=True,
                      nrows=int(u.shape[1]/3),
                      names=[0,1,2,'x','y','z'],
                      usecols=[2,'x','y','z'],
                      index_col=0)
uheader.index.name=u.columns.names[0]

dt = np.mean(np.diff(u.index))
fs = 1/dt
uStar = 0.41*10/np.log(1/0.00003)
nfft=512
from scipy.signal import welch
n, Su = welch(up[0], fs, axis=0, nperseg=nfft)
Su = pd.DataFrame(Su, columns=up[0].columns, index=n)
ndSu = Su.apply(lambda x: x*n/uStar**2)
print(ndSu)
ndSu.index = n*uheader.z[0]/u_orig[0, 'ux'].mean()
f = ndSu.index.values
ndSu['Kaimal u'] = 200.*f/(1+50.*f)**(5./3.)
ndSu['Kaimal v'] = 15.*f/(1+9.5*f)**(5./3.)
ndSu['Kaimal w'] = 3.36*f/(1+10.*f**(5./3.))
styles = ['ks','ro','gv','k','r','g']
ax = ndSu.plot(loglog=True,style=styles)
ax.set_xlabel('$nz/\\langle u\\rangle$')
ax.set_ylabel('$nS_u/u_*^2$')
ax.legend(['u','v','w'], loc='best')
#ax.set_xlim(0.005,10)
#ax.set_ylim(10e-4, 10e1)
fig=plt.gcf()
fig.tight_layout()
#fig.savefig('{0}psd_u{1}.eps'.format(figPath,point))
plt.show()
```
